```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# matplot plot

```python
day=[1,2,3,4,5,6]
temp=[28,30,34,25,28,30]
plt.plot(day,temp,marker='o',linestyle="--",color="red")
```

```python
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Climate Change")
plt.plot(day,temp,marker='o',linestyle="dotted",color="green",linewidth=5)
```

# matlib bar

```python
import numpy as np
company=["Reliancce","Tata","Adani"]
reve=[90,80,95]
xpos=np.arange(len(company))
plt.bar(xpos,reve)
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title("Indian Companies")
plt.bar(xpos,reve,width=0.2,color="green",label="Revenue")
plt.legend()
plt.show()
```

```python
profit=[40,30,35]
plt.xticks(xpos,company)
plt.ylabel("Revenue with profit")
plt.title("Indian companies")
plt.bar(xpos-0.2,reve,width=0.4,color="gray",label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,color="black",label="Profit")
```

# horizontal bar

```python
plt.yticks(xpos,company)
plt.xlabel("Revenue")
plt.barh(xpos,reve,height=0.2,color="orange",label="Revenue")
plt.legend()
plt.show()
```

# pie chart

```python
mobile=['apple','oppo','one1','vivo','MI']
shares=[350,700,1200,600,200]
plt.pie(shares,labels=mobile)
```

```python
plt.pie(shares,labels=mobile,shadow=True)
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%',radius=1.2)
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%',radius=1.2,startangle=0,explode=[0.1,0,0,0,0])
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%',radius=1.2,startangle=0,explode=[0.1,0,0,0,0],counterclock=True)
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%',radius=1.2,startangle=0,explode=[0.1,0,0,0,0],counterclock=False)
plt.show()
```

```python
plt.pie(shares,labels=mobile,shadow=True,autopct='%1.1f%%',radius=1.2,startangle=0,explode=[0.1,0,0,0,0],counterclock=False)
plt.legend(loc="be")
plt.show()
```
